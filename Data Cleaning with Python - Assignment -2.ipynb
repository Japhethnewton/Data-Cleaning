and inaccuracies in datasets. It involves a series of steps such as data validation, data
transformation, data normalization, and data enrichment to ensure that the data is accurate,

a critical step in data preprocessing and is often performed before data analysis or machine
learning algorithms are applied to the data.

# Task 1 - Where Name is missing fill the value in place by extracting the name from the Email-id column.

```python
# Add values to NaN values that are missing with cetain names from another row email
Um['Name'] = Um['Name'].fillna(Um['Email'].str.split('@').str[0])
Um['Name'] = Um['Name'].str.replace(".", "")

# using the split abd replace meathod to remove (".") and the split the eamil ("@") for it to extracted to Name column
```

```python
# Calculate the mean of the Age column
mean_age = Um['Age'].mean()

print(mean_age)

```

```python
# Filling missing age values with the mean
Um['Age'].fillna(mean_age, inplace=True)
```

# Task 5 - Highlight the Top 10 Employees with highest salary in the dataset.Convert 'Salary' column to numeric type



```python
# Convert 'Salary' column to numeric type
Um['Salary'] = pd.to_numeric(Um['Salary'])
```

```python
# Get the top 10 salaries
top_10_salaries = Um.nlargest(10, 'Salary')
```

```python
# making Function to apply highy light color to the Name column and salary column
def highlight_name(name):
    color = 'blue' if name in top_10_salaries['Name'].values else 'black'
    return f'color: {color}'

# Function to apply color to the Salary column
def highlight_salary(salary):
    color = 'red' if salary in top_10_salaries['Salary'].values else 'black'
    return f'color: {color}'

# Apply the color highlighting to the Name and Salary with the top 10 Employee_Highest_Salary
highlighted_to_10_Employee_Highest_Salary = Um.style.applymap(highlight_name, subset=['Name']).applymap(highlight_salary, subset=['Salary'])
```

# Task 6 - Where Name is there but no Email-Id then fill the blank with (support@dataisgood.com)

# Task 7-In the Department column make the data proper with the useful functions and rename the wrong spelled words in the department column.


```python
# Remove leading and trailing spaces in department names
Um['Department'] = Um['Department'].str.strip()

```

```python
# Replace misspelled words in department names
corrections = {
    'DeVlopemENT': 'Development',
    'SciENce': 'Science',
    'Mangement': 'Management',
    'OperarYION': 'Operation',
    'SALES' : 'Sales','SLES' :'Sales', 'DaTa Science' : 'Data Science' , 'HR' : 'Hr' , 'Finance/Accounts' : 'Finance Accounts',
     
}
Um['Department'] = Um['Department'].replace(corrections, regex=True)
```
